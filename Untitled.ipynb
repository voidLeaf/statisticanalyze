```r
library('MASS')
library('sampling') 
d<-read.csv('./data2.txt',fileEncoding="UTF-8");
d<-data.frame(d)
ls<-d[31,]
d<-d[-31,]
d.comp<-scale(d[,3:12], center=T,scale=T)
d.pr<-princomp(d.comp,cor=F)
```

```r
summary(d.pr,loadings = T)
```

```r
screeplot(d.pr,type="lines")
cor(d.comp)
```

```r
d.newComp =  d.comp%*%d.pr$loadings[,1:5]
d.hc <- hclust( dist(d.newComp),method = "complete",members=NULL)
plot(d.hc,hang=-1,labels=d[,2])
rect.hclust(d.hc, k = 5)
d$hcclass <- cutree(d.hc, 5)
```

```r
classCount <- 5

d.km <- kmeans(d.newComp, classCount)
d$kmclass<-d.km$cluster
d[,c(1,13)][order(d$hcclass),]
```

```r
d[order(d$hcclass),][c(1,13)]
d[order(d$hcclass),]$hcclass
d[order(d$hcclass),]$地区
```
